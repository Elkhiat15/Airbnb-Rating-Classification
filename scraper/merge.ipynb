```python
import pandas as pd

KAGGLE_DATASET_PATH = "Airbnb_Data.csv"
SCRAPED_DATASET_PATH = "airbnb_listings_.csv"
```

```python
kaggle_df = pd.read_csv(KAGGLE_DATASET_PATH)
kaggle_df.head()
```

```python
scraped_df = pd.read_csv(SCRAPED_DATASET_PATH)
scraped_df.head()
```

```python
scraped_df.info()
```

```python
kaggle_df.info()
```

```python
import pandas as pd
import numpy as np

# Extract ID from URL
scraped_df['id'] = scraped_df['listing_url'].str.extract(r'/rooms/(\d+)')[0].astype(np.int64)

# Drop unnecessary columns
scraped_df = scraped_df.drop(columns=['listing_url', 'is_superhost'])

# Transform free_cancellation to match Kaggle's cancellation_policy
# We'll convert True/False to 'flexible'/'strict' for example (or 'Yes'/'No' as a placeholder)
scraped_df['free_cancellation'] = scraped_df['free_cancellation'].map({True: 'Yes', False: 'No'})

# Rename columns to match Kaggle
scraped_df = scraped_df.rename(columns={
    'guests': 'accommodates',
    'rating': 'review_scores_rating',
    'review_count': 'number_of_reviews',
    'free_cancellation': 'cancellation_policy',
    'price_per_night': 'price'
})
```

```python
# Drop unnecessary columns
kaggle_df = kaggle_df.drop(columns=[
    'bed_type', 'cleaning_fee', 'description', 'first_review',
    'host_has_profile_pic', 'host_since', 'instant_bookable', 'last_review',
    'name', 'thumbnail_url', 'zipcode'
])

# Transform log_price back to linear scale
kaggle_df['price'] = np.exp(kaggle_df['log_price'])
kaggle_df = kaggle_df.drop(columns=['log_price'])

# Scale review_scores_rating from 0-100 to 0-5
kaggle_df['review_scores_rating'] = kaggle_df['review_scores_rating'] / 20
```

```python
# Get Kaggle columns after transformation
columns_order = kaggle_df.columns.tolist()

# Reorder scraped_df to match Kaggle
scraped_df = scraped_df[columns_order]
```

```python
merged_df = pd.concat([kaggle_df, scraped_df], ignore_index=True)
print(merged_df.info())
```

```python
def show_unique(column):
    kaggle_unique_values = kaggle_df[column].unique()
    scraped_unique_values = scraped_df[column].unique()
    print(f"Unique values in '{column}':")
    print(f"\nFOR KAGGLE DATASET: {kaggle_unique_values}")
    print(f"{30 * '-'}")
    print(f"\nFOR SCRAPED DATASET: {scraped_unique_values}\n")
```

```python
attributes = ['room_type', 'city', 'host_identity_verified']
for attr in attributes:
    print(f"{60 * '#'}\n")
    show_unique(attr)
```
